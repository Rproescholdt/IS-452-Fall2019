# While loops

When it comes to data processing in python, the while loop is only deployed when the for loop won't cut it.  

While loops are the standard way to loop for some languages, but not so much in python.  There are certain tasks, such as interactive programs, where the while loop is almost always what you use.

This class is dealing with data, which is usually in files so you know exactly how many times to run.

The while loop is valuable in the cases when you don't know how many times you'll need to run through things.

Sometimes we might know how many times we want to do something, but not how many times that thing will take.  In this way, we can combine our two patterns:

1. a for loop to just make stuff go a certain number of times
2. a while loop inside that for loop to get however much data that we want.

```python
fileio = open('crappystatedata.txt', 'r')

text = fileio.read()
lines = text.split("\n")

allstatechunks = []

pos_index = 0
counter = 0

while counter < 51:
    current_line = lines[pos_index]
    
    # the nested loop will push the index up
    state = current_line
    print(state)
    
    pos_index += 1
    current_line = lines[pos_index]
    
    # establish blank state chunk
    statedata = []
    
    # add state line
    statedata.append(state)
    
    print(pos_index, len(lines))
    while current_line.endswith('[edit]') == False and pos_index < len(lines) -1:
        statedata.append(current_line)
        pos_index += 1
        current_line = lines[pos_index]

    allstatechunks.append(statedata)
    
    
fileio.close()
```

```python
fileio = open('crappystatedata.txt', 'r')

text = fileio.read()
lines = text.split("\n")

allchunks = []
statechunk = []

foundfirst = False

for line in lines:
    if line.endswith("[edit]") and foundfirst == False:
        print("This is the first state!", line)
        statechunk.append(line)
        foundfirst = True
    elif line.endswith("[edit]"):
        print("a new state has begun!", line)
        # add completed state
        allchunks.append(statechunk)
        # reset the chunk
        statechunk = [line]
    else:
        statechunk.append(line)
    allchunks.append(statechunk)
        
```
